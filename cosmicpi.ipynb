ABAIXO ESTÁ O CÓDIGO EM PYTHON PARA A ESTIMATIVA DA TAXA DE MUONS INCIDENTES NO COSMICPI PARA x=e, levando em consideração a eficiência de detecção de múons e a versão do detector.

```
x = int(input(print("entre com o número de eventos:")))
  
def prob(x):                                                                     
  fat = 1
  i = 1
  while (i <= int(x)):
    fat = fat*i
    i = i + 1
  e = 2.71828
  m = (100*200*(1/(60*100)))*1                           #de acordo com os dados coletados, eficiencia = 1 é o que melhor se ajusta aos resultados
  p = (e**(-m))*(m**x)/fat
  return(p)
  print(p)

print("Para o caso de", x, "eventos:", prob(x))
```

ABAIXO ESTÁ O CÓDIGO EM PYTHON PARA A ESTIMATIVA DA TAXA DE MUONS INCIDENTES NO COSMICPI PARA x<=e, levando em consideração a eficiência de detecção de múons e a versão do detector.

```
def fatorial(x):
    l = []
    fatorial = 1
    for i in range(1, x + 1):
        while (i > 0):
            fatorial *= i
            i -= 1
        l.append(float(fatorial))
        fatorial = 1
    return l


def prob(e, m, cum, l):
    print("Para o caso de 0 eventos:",((e**(-m))*(m**0)))
    for cont in range(len(l)):
        cum += ((e**(-m))*(m**(cont+1)))/l[cont]
        print("Para o caso de", cont+1, "eventos:",cum)


def main():
    x = int(input(print("entre com o número de eventos:")))
    cum, e = 0.0, 2.71828
    m = (100*200*(1/(60*100)))*1                        #de acordo com os dados coletados, eficiencia = 1 é o que melhor se ajusta aos resultados
    l = fatorial(x)
    prob(e, m, cum, l)


main()
```

```
from matplotlib import pyplot as plt
from scipy.stats import poisson
import numpy as np
from google.colab import drive 
drive.mount('/content/gdrive') 

x = 10
m = (100*200*(1/(60*100)))*1                  #de acordo com os dados coletados, eficiencia = 1 é o que melhor se ajusta aos resultados 

a = np.arange(x)
print(poisson.cdf(a,m))
print(poisson.pmf(a,m))

fig, ax = plt.subplots(figsize=(len(a), len(poisson.cdf(a, m))), dpi=100)
ax.plot(a,poisson.cdf(a, m), poisson.pmf(a,m), "k-")
ax.grid(which="both")
ax.set_title("Estimativa da taxa de múons")
ax.set_xlabel("Número de eventos por segundo")
ax.set_ylabel("Probabilidade")
plt.savefig(fname = "cosmicpi.png", edgecolor = "white", bbox_inches = "tight", dpi=fig.dpi) 
plt.show()
```
